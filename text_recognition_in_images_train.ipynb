### iniciando

```python
import pytesseract
import cv2
from google.colab.patches import cv2_imshow
```

```python
img = cv2.imread('/content/drive/MyDrive/Visão Computacional Guia Completo/Images/saida.jpg')
cv2_imshow(img)
```

```python
config_tesseract = '--tessdata-dir tessdata --psm 7'
text = pytesseract.image_to_string(img, lang='por', config=config_tesseract)
print(text)
```
